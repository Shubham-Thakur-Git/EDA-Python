**Log transformation**

```python
# Natural logarithm, element-wise
# 
The natural logarithm log is the inverse of the exponential function, so that log(expx))( =x  #  The natural logarithm is logarith in base e.
```

```python
x = 10
np.log(10)     # This is base e
```

```python
log_transform = np.log(exp_data)
log_transform
```

```python
sns.histplot(log_transform,bins=20,kde=True)
plt.show()
```

```python
plt.hist(exp_data,bins=20,label='Exponential Data')
plt.hist(log_transform,bins=20,label='Log Transformed Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(log_transform,bins=20,label='Log Transformed Data')
plt.legend()

plt.subplot(1,2,2)
plt.hist(exp_data,bins=20,label='Exponential Data')
plt.legend()

plt.show()
```

**Sqrt Transformation**

```python
sqrt_transform = np.sqrt(exp_data)
sqrt_transform
```

```python
sns.histplot(data=sqrt_transform,bins=20,kde=True)
plt.show()
```

```python
sns.kdeplot(data=sqrt_transform)
plt.show()
```

```python
plt.hist(exp_data,bins=20,label='Exponential Data')
plt.hist(sqrt_transform,bins=20,label='Sqrt Transformed Data')
plt.legend()
plt.show()
```

**Reciprocal Transformation**

```python
rec_transform = np.reciprocal(exp_data)
rec_transform
```

```python
sns.histplot(rec_transform,bins=20,kde=True)
plt.show()
```

**Power Transformation**

```python
from sklearn.preprocessing import PowerTransformer
PT = PowerTransformer()
visa_df['PT_wage_data'] = PT.fit_transform(visa_df[['prevailing_wage']])
#power_transform = PT.fit_transform(exp_data)
```

```python
plt.hist(visa_df['PT_wage_data'],bins=20)
plt.show()
```

```python
visa_df['PT_wage_data'].max()
```

```python
visa_df['PT_wage_data'].min()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.hist(visa_df['prevailing_wage'],bins=20,label='Prevailing Wage')
plt.legend()

plt.subplot(1,2,2)
plt.hist(visa_df['PT_wage_data'],bins=20,label='Power Transform Wage')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.boxplot(visa_df['prevailing_wage'],label='Prevailing Wage')
plt.legend()

plt.subplot(1,2,2)
plt.boxplot(visa_df['PT_wage_data'],label='Power Transform wage')
plt.legend(loc='upper right')

plt.show()
```

- Transformations affect the outliers of the data
    - Log Transformation
    - Exp Transformation
    - Reciprocal Transformation
    - Sqrt Transformation
    - Power Transformation

- Scaling of the data
    - Z Scale
    - Min max scale
 

- Two approaches:
    - First treat the outliers then do transformation or scaling
    - Second, Transform or scale the data, then work on outliers
        - In this case, if you still found any outliers, it is called as **influential outliers**

- We cannot say which is the best method, it completely depends on problem to problem
